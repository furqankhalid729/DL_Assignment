```python
import numpy as np
import glob
from PIL import Image
import pandas as pd
from keras import layers,models
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
train_ds =keras.preprocessing.image_dataset_from_directory(
    directory='/content/drive/MyDrive/Kaggle Dataset/flowers/',
    labels='inferred',
    label_mode='categorical',
    shuffle=True,
    validation_split=0.3,
    subset="training",
    batch_size=32,
    seed=111,
    image_size=(180, 180)
)

test_ds =keras.preprocessing.image_dataset_from_directory(
    directory='/content/drive/MyDrive/Kaggle Dataset/flowers/',
    labels='inferred',
    label_mode='categorical',
    shuffle=True,
    validation_split=0.3,
    subset="validation",
    batch_size=32,
    seed=111,
    image_size=(180, 180)
)

```

```python
train_ds
```

```python
%tensorflow_version 2.x
import tensorflow as tf
```

```python
model = keras.applications.Xception(weights=None, input_shape=(180, 180, 3), classes=5)
model.compile(optimizer='rmsprop', loss='categorical_crossentropy')
```

```python
with tf.device('/device:GPU:0'):  
  model.fit(train_ds, epochs=10, validation_data=test_ds,use_multiprocessing=True)
```

```python
model.evaluate(test_ds)
```
